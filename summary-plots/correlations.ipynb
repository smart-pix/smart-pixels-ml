```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import matplotlib.colors as mcolors
colors = list(mcolors.TABLEAU_COLORS.keys())
```

```python
df = pd.read_parquet('max_2dconv_noquant_20t.parquet')
```

```python
fig, ax = plt.subplots(3,3,figsize=(6,6),sharey=True,constrained_layout=True)

bins = np.linspace(-1,1,20)

plt.yscale('log')

ax[0,0].set_ylabel('Clusters')
ax[1,0].set_ylabel('Clusters')
ax[2,0].set_ylabel('Clusters')
plt.ylim(0.5,5*10**5)

ax[0,0].set_xlabel(r'cov$(x,y)/xy$')
ax[1,0].set_xlabel(r'cov$(x,\cot\alpha)/x\cot\alpha$')
ax[2,0].set_xlabel(r'cov$(x,\cot\beta)/x\cot\beta$')
ax[1,1].set_xlabel(r'cov$(y,\cot\alpha)/y\cot\alpha$')
ax[2,1].set_xlabel(r'cov$(y,\cot\beta)/y\cot\beta$')
ax[2,2].set_xlabel(r'cov$(\cot\alpha,\cot\beta)/\cot\alpha\cot\beta$')

ax[0,0].hist(df['cov_xy']/(df['x']*df['y']),histtype='step',bins=bins)
ax[1,0].hist(df['cov_xcotA']/(df['x']*df['cotA']),histtype='step',bins=bins)
ax[2,0].hist(df['cov_xcotB']/(df['x']*df['cotB']),histtype='step',bins=bins)
ax[1,1].hist(df['cov_ycotA']/(df['y']*df['cotA']),histtype='step',bins=bins)
ax[2,1].hist(df['cov_ycotB']/(df['y']*df['cotB']),histtype='step',bins=bins)
ax[2,2].hist(df['cov_cotAcotB']/(df['cotA']*df['cotB']),histtype='step',bins=bins)

ax[0,1].axis('off')
ax[0,2].axis('off')
ax[1,2].axis('off')

plt.tight_layout()

plt.savefig("plots/cov.pdf")
plt.savefig("plots/cov.png")
```
