```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 12})
```

```python
df_2s_test = pd.DataFrame()
path_test = "/eos/uscms/store/user/jennetd/smart-pixels/parquet/shuffled/dataset_2s_16x16_50x12P5_centeredIncidence_parquets/test/"

cols = ['x-midplane','y-midplane','cotAlpha','cotBeta','original_atEdge']

for f in os.listdir(path_test):
    this_df = pd.read_parquet(path_test+f,columns=cols)
    df_2s_test = pd.concat([df_2s_test,this_df[this_df['original_atEdge']< 1]])

print("testing dataset = " + str(len(df_2s_test)))
```

```python
df_2s_train = pd.DataFrame()
path_train = "/eos/uscms/store/user/jennetd/smart-pixels/parquet/shuffled/dataset_2s_16x16_50x12P5_centeredIncidence_parquets/train/"

cols = ['x-midplane','y-midplane','cotAlpha','cotBeta','original_atEdge']

for f in os.listdir(path_train):
    this_df = pd.read_parquet(path_train+f,columns=cols)
    df_2s_train = pd.concat([df_2s_train,this_df[this_df['original_atEdge']< 1]])

print("training dataset = " + str(len(df_2s_test)))
```

```python
df_3sr_test = pd.DataFrame()
path_test = "/eos/uscms/store/user/jennetd/smart-pixels/parquet/shuffled/dataset_3sr_16x16_50x12P5_centeredIncidence_parquets/test/"

cols = ['x-midplane','y-midplane','cotAlpha','cotBeta','original_atEdge']

for f in os.listdir(path_test):
    this_df = pd.read_parquet(path_test+f,columns=cols)
    df_3sr_test = pd.concat([df_3sr_test,this_df[this_df['original_atEdge']< 1]])

print("testing dataset = " + str(len(df_3sr_test)))
```

```python
df_3sr_train = pd.DataFrame()
path_train = "/eos/uscms/store/user/jennetd/smart-pixels/parquet/shuffled/dataset_3sr_16x16_50x12P5_centeredIncidence_parquets/train/"

cols = ['x-midplane','y-midplane','cotAlpha','cotBeta','original_atEdge']

for f in os.listdir(path_train):
    this_df = pd.read_parquet(path_train+f,columns=cols)
    df_3sr_train = pd.concat([df_3sr_train,this_df[this_df['original_atEdge']< 1]])

print("training dataset = " + str(len(df_3sr_train)))
```

```python
fig, ax = plt.subplots(1,4,figsize=(12,3),sharey=True,constrained_layout=True)

ax[0].hist(df_3sr_train['x-midplane'] - 50./2,weights=[1./len(df_3sr_train)]*len(df_3sr_train),histtype='step')
ax[0].set_ylabel('Fraction of clusters')
ax[0].set_xlabel('True x [um]')
ax[0].set_xlim(-125,125)
ax[1].hist(df_3sr_train['y-midplane'] - 12.5/2,weights=[1./len(df_3sr_train)]*len(df_3sr_train),histtype='step')
ax[1].set_xlabel('True y [um]')
ax[1].set_xlim(-30,30)
ax[2].hist(df_3sr_train['cotAlpha'],weights=[1./len(df_3sr_train)]*len(df_3sr_train),histtype='step')
ax[2].set_xlabel(r'True $\cot\alpha$')
ax[2].set_xlim(-8,8)
ax[3].hist(df_3sr_train['cotBeta'],weights=[1./len(df_3sr_train)]*len(df_3sr_train),histtype='step')
ax[3].set_xlabel(r'True $\cot\beta$')
ax[3].set_xlim(-2.5,2.5)

#ax[0].hist(df_test['x-midplane']-50./2,weights=[1./len(df_test)]*len(df_test),histtype='step')
#ax[1].hist(df_test['y-midplane']-12.5/2,weights=[1./len(df_test)]*len(df_test),histtype='step')
#ax[2].hist(df_test['cotAlpha'],weights=[1./len(df_test)]*len(df_test),histtype='step')
#ax[3].hist(df_test['cotBeta'],weights=[1./len(df_test)]*len(df_test),histtype='step')

fig.savefig('plots/truth_3sr.png',bbox_inches='tight')
fig.savefig('plots/truth_3sr.pdf',bbox_inches='tight')
plt.show() 
```
